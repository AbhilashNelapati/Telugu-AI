```python
import streamlit as st
from googletrans import Translator
from gtts import gTTS
import torch
from transformers import pipeline
from pydub import AudioSegment
import numpy as np
from scipy.io import wavfile
from io import BytesIO







```

```python
import sounddevice as sd
import numpy as np
import wavio
import io
```

```python

def record_audio():
    fs = 44100  # Sampling frequency
    duration = 10  # Duration of recording in seconds
    silence_limit = 10  # Silence limit in seconds
    audio_data = []

    def callback(indata, frames, time, status):
        if status:
            print(status, flush=True)
        if np.any(indata):
            audio_data.append(indata.copy())

    with sd.InputStream(samplerate=fs, channels=1, callback=callback):
        print("Recording...")
        sd.sleep(duration * 1000)
        print("Recording stopped.")

    if not audio_data:
        return None

    audio_data = np.concatenate(audio_data, axis=0)
    audio_buffer = io.BytesIO()
    wavio.write(audio_buffer, audio_data, fs, sampwidth=2)
    audio_buffer.seek(0)
    return audio_buffer.read()
```

```python
 # Title for the Streamlit app
st.title("Telugu ASR and Translation App")
# ASR Section
audio = "/content/output.mp3"
device = "cuda:0" if torch.cuda.is_available() else "cpu"
model_name = "vasista22/whisper-telugu-base"

transcribe = pipeline(
    task="automatic-speech-recognition",
    model=model_name,
    chunk_length_s=30,
    device=device
)
```

```python


transcribe.model.config.forced_decoder_ids = transcribe.tokenizer.get_decoder_prompt_ids(language="te", task="transcribe")
transcription = transcribe(audio)["text"]





# Display ASR Output (Transcription) in Streamlit
st.write('ASR Output (Transcription): ', transcription)
```

```python
print('ASR Output (Transcription): ', transcription)
```

```python

# TTS Section
tts = gTTS(text=transcription, lang='te')
output_file_te = "translated_output.mp3"
tts.save(output_file_te)


#diplay the telugu audio
display(Audio(output_file_te, autoplay=True)    )

# Provide audio playback in Streamlit
st.audio(output_file_te, format='audio/mp3')
```

```python


# Translate Telugu text to English
translator = Translator()
translated = translator.translate(transcription, src='te', dest='en')
translated_text = translated.text

# Display Translated Text in Streamlit
st.write('Translated Text: ', translated_text)


print('Translated Text: ', translated_text)
```

```python

# TTS Section
tts = gTTS(text=translated_text, lang='en')
output_file_en = "translated_output.mp3"
tts.save(output_file_en)

      # Provide audio playback in Streamlit
st.audio(output_file_en, format='audio/mp3')
```
